# 2.1 以 DataFrame 格式載入 train.csv (5%)

```python
import pandas as pd
titanic_df = pd.read_csv('train.csv')
titanic_df.head()
```

# 2.2 請檢查並列出 train.csv 中每個欄位的空值個數 (5%)

```python
titanic_df.isnull().sum()
```

# 2.3 將 Age 欄位空值以該性別平均值填入。(10%)

```python
print("male mean age : ", titanic_df[titanic_df.Sex=='male'].Age.mean())
print("female mean age : ", titanic_df[titanic_df.Sex=='female'].Age.mean())
```

```python
titanic_df['Age'] = titanic_df.groupby('Sex')['Age'].apply(lambda x : x.fillna(x.mean()))
titanic_df
```

```python
print("male mean age : ", titanic_df[titanic_df.Sex=='male'].Age.mean())
print("female mean age : ", titanic_df[titanic_df.Sex=='female'].Age.mean())
print("missing value count: ", pd.isnull(titanic_df.Age).sum())
```

# 2.4 將 Cabin 欄位重製為 Pclass * Fare (5%)

```python
titanic_df['Cabin'] = titanic_df.Pclass * titanic_df.Fare
titanic_df
```

# 2.5 將 Survived 欄位重製為 0=Alive 1=Dead (5%)

```python
titanic_df['Survived'] = titanic_df['Survived'].replace(to_replace=0, value='Alive')
titanic_df['Survived'] = titanic_df['Survived'].replace(to_replace=1, value='Dead')
titanic_df
```

# 2.6 將 Sex 與 Embarked 欄位轉為數字型態（例如：男性=0，女性=1）

```python
titanic_df['Sex'] = titanic_df['Sex'].replace(to_replace='male', value=0)
titanic_df['Sex'] = titanic_df['Sex'].replace(to_replace='female', value=1)
```

```python
titanic_df['Embarked'] = titanic_df['Embarked'].replace(to_replace='S', value=0)
titanic_df['Embarked'] = titanic_df['Embarked'].replace(to_replace='C', value=1)
titanic_df['Embarked'] = titanic_df['Embarked'].replace(to_replace='Q', value=2)
```

```python
titanic_df
```

## 處理剩下的missing value

```python
titanic_df.isnull().sum()
```

```python
titanic_df = titanic_df.dropna()
```

```python
titanic_df.isnull().sum()
```

# 2.7 請以 PassengerId、Survived、Name、Ticket、Pclass 以外的欄位作為訓練資料，建立 Decision tree 來預測 Survived，將訓練資料比例設為50%，random_state 設為 12，stratify = y，並繪出 Decision tree 的樹狀圖 (10%)

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import classification_report
```

```python
unselectd_columns = ['PassengerId', 'Survived', 'Name', 'Ticket', 'Pclass']
selected_columns = [col for col in titanic_df if col not in unselectd_columns]
selected_columns
```

```python
X = np.array(titanic_df[selected_columns])
y = np.array(titanic_df['Survived'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=12, stratify=y)
```

```python
model = DecisionTreeClassifier(random_state=12)
model = model.fit(X_train, y_train)
```

# 2.8 計算出在2.7測試資料上的平均準確率 (5%)

```python
# test data accuracy
model.score(X_test, y_test)
```

```python
# traini data accuracy
model.score(X_train, y_train)
```

# 2.9 請用2.7的結果評估決策樹好壞(使用classification_report)產生類似以下結果(5%)

```python
pred = model.predict(X_test)
```

```python
target_names = ['Alive', 'Dead']
print(classification_report(y_test, pred, target_names=target_names))
```

# 2.10 請分別以訓練資料比例 60%、70%、80%、90% 建立 Decision tree，random_state皆設為 12，並將不同資料比例與平均準確率的比較結果以 DataFrame 呈現，如右圖所示。 (10%)

```python
def outputTreeModelResult(X, y , trainRatio):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=trainRatio, random_state=12, stratify=y)
    model = DecisionTreeClassifier(random_state=12)
    model = model.fit(X_train, y_train)

    return model.score(X_test, y_test)
```

```python
trainRationList = ['50/50', '60/40', '70/30', '80/20', '90/10']
outResList = [outputTreeModelResult(X, y, 0.5), outputTreeModelResult(X, y, 0.6), outputTreeModelResult(X, y, 0.7), 
             outputTreeModelResult(X, y, 0.8), outputTreeModelResult(X, y, 0.9)]

test_df = pd.DataFrame({'split_proportion' : trainRationList, 'score' : outResList})
test_df
```

# 2.11呈上題，將此比較結果以折線圖呈現，如下圖所示：(5%)

```python
import matplotlib.pyplot as plt

plt.plot(trainRationList, outResList)
plt.title('train/test split proportion & result')
plt.xlabel('train/test proportion')
plt.legend(['score'])
plt.show()
```
