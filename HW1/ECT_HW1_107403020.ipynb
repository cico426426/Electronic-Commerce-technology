## 處理Product占用欄位 function

```python
order_list = ['E-mail', 'Fax', 'Mail', 'Sales', 'Special', 'Telephone', 'Web']
```

```python
#其餘數值名稱
other_list = ['Revenue_', 'Planned revenue', 'Product cost', 'Quantity_', 'Unit cost', 'Unit price', 'Gross profit', 'Unit sale price']
```

```python
#根據 product 占用其他column的情形，將其分類
def classified_product_by_occupied_columns(product, columns):
    product_ = sales_df[sales_df[columns].isin(order_list)].copy(deep=True)
    product_.groupby(['Product.2', columns]).size()
    print(f"{product}, {len(product_.index)}")
    return product_

# 將合併 product 會用到的 columns 整理出來, offset 為占用了多少column 
def creatre_product_removed_slots(df, offset):
    return df.columns[1:1+offset]

# 回傳合併好的一系列 product
def get_product_name(df, slots):
    value = ''
    for slot in slots:
        df = df.astype({slot:str})
        value = value + df[slot] + ' '
    return value
```

# 9. 請檢查第 5個欄位 Order method type的資料錯置情況並將此欄位整理成正確的形式。(1 5%)

```python
#根據 order 占用其他column的情形，將其分類
def classified_order_by_occupied_columns(df):
    order1 = df[df[df.columns[1]]!='Sales'].copy(deep=True)
    order2 = df[df[df.columns[1]]=='Sales'].copy(deep=True)
    return order1, order2

# 處理 order1 
def modify_order1(df):
    order_slot = df.columns[1]
    df['order method type'] = df[order_slot]
    df = df.drop(order_slot, axis=1)
    return df

# 處理 order2
def modify_order2(df):
    order_slots = df.columns[1:3]
    df['order method type'] = df[order_slots[0]] + ' ' + df[order_slots[1]]
    df = df.drop(order_slots, axis=1)
    return df
```

# 10. 請檢查第 6 個欄位 Retailer country 的資料錯置情況並將此欄位整理成正確的形式。 (15%)

```python
#根據 country 占用其他column的情形，將其分類
def classified_country_by_occupied_columns(df):
    country_2wd_idx = df.columns[2]
    country_2wd = [product for product in df[country_2wd_idx].unique() if isinstance(product, str)] #占用2格columns
    country1 = df[~df[country_2wd_idx].isin(country_2wd)].copy(deep=True)
    country2 = df[df[country_2wd_idx].isin(country_2wd)].copy(deep=True)
    return country1, country2

# 處理 country1
def modify_country1(df):
    country_slot = df.columns[1]
    df['Retailer country'] = df[country_slot]
    df = df.drop(country_slot, axis=1)
    return df

# 處理 country2
def modify_country2(df):
    country_slots = df.columns[1:3]
    df['Retailer country'] = df[country_slots[0]] + ' ' + df[country_slots[1]]
    df = df.drop(country_slots, axis=1)
    return df
```

# 11. 請將剩下的欄位整理成正確的形式。(10%)

```python
# 整理剩下且為數值的columns
def create_numerical_columns(df):
    return df.columns[1: df.columns.get_loc('Product line')]

# 處理country佔用一欄時，剩下的數值columns
def modify_country1_numerical_columns(df):
    numerical_columns = create_numerical_columns(df)
    for i in range(len(other_list)):
        df[other_list[i]] = df[numerical_columns[i]]
    for col in numerical_columns:
        df = df.drop([col], axis=1)
    df = df.rename(columns={'Revenue_' : 'Revenue'})
    df = df.rename(columns={'Quantity_' : 'Quantity'})
    return df

# 處理country佔用倆欄時，剩下的數值columns
def modify_country2_numerical_columns(df):
    numerical_columns = create_numerical_columns(df)
    for i in range(len(other_list)):
        df[other_list[i]] = df[numerical_columns[i]]
    for col in numerical_columns:
        df = df.drop([col], axis=1)
    df = df.rename(columns={'Revenue_' : 'Revenue'})
    df = df.rename(columns={'Quantity_' : 'Quantity'})
    return df
```

# 8~11 步驟合併

```python
def modify_product_n(df, n):
    prduct_slots = creatre_product_removed_slots(df, n)
    df['Product'] = get_product_name(df, prduct_slots)
    df = df.drop(prduct_slots, axis=1)
    return df

def deal_order(product):
    o1, o2 = classified_order_by_occupied_columns(product)
    o1 = modify_order1(o1)
    o2 = modify_order2(o2)
    o1_c1, o1_c2 = deal_country(o1)
    o2_c1, o2_c2 = deal_country(o2)
    return o1_c1, o1_c2, o2_c1, o2_c2

def deal_country(order):
    c1, c2 = classified_country_by_occupied_columns(order)
    c1 = modify_country1(c1)
    c2 = modify_country2(c2)
    return modify_country1_numerical_columns(c1), modify_country2_numerical_columns(c2)

def concat_product(o1, o2, o3, o4):
    return pd.concat([o1, o2, o3, o4])

# 根據product占用欄位，處理product
def deal_product_n(df, n):
    df = modify_product_n(df, n)
    p_o1_c1, p_o1_c2, p_o2_c1, p_o2_c2 = deal_order(df)
    df = concat_product(p_o1_c1, p_o1_c2, p_o2_c1, p_o2_c2)
    return df
```

# 分類product

```python
product_1 = classified_product_by_occupied_columns(1, 'Order')
product_2 = classified_product_by_occupied_columns(2, 'method')
product_3 = classified_product_by_occupied_columns(3, 'type.1')
product_4 = classified_product_by_occupied_columns(4, 'Retailer')
product_5 = classified_product_by_occupied_columns(5, 'country')
product_6 = classified_product_by_occupied_columns(6, 'Revenue')
product_7 = classified_product_by_occupied_columns(7, 'Planned')
```

# 個別處理各個product

```python
product_1 = deal_product_n(product_1, 1)
product_2 = deal_product_n(product_2, 2)
product_3 = deal_product_n(product_3, 3)
product_4 = deal_product_n(product_4, 4)
product_5 = deal_product_n(product_5, 5)
product_6 = deal_product_n(product_6, 6)
```

# 合併處理完畢的product

```python
product_df = pd.concat([product_1, product_2, product_3, product_4, product_5, product_6])
product_df.head()
```
