```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 1. 載入 Churn_Modelling.csv 資料集，並印出哪些欄位含有遺漏值(missing value)。 (5%)

```python
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

# 2. 以平均值填入 EstimatedSalary 的遺漏值，以眾數填入 Age 與 Gender 的遺漏值。(10%)

```python
df['EstimatedSalary'] = df['EstimatedSalary'].fillna(df['EstimatedSalary'].mean())
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df.isnull().sum()
```

# 3. 修改欄位名稱，將 CredRate 改成 CreditScore、ActMem 改成IsActiveMember、Prod Number 改成 NumOfProducts、Exited 改成Churn，以利後續分析資料。 (5%)

```python
df.rename(columns={'CredRate' : 'CreditScore', 
           'ActMem' : 'IsActiveMember', 
           'Prod Number' : 'NumOfProducts', 
           'Exited' : 'Churn'}, inplace=True)
```

```python
df.head()
```

# 4. 去除 CustomerId,欄位，並將Geography、Gender、HasCrCard、 Churn、IsActiveMember 修改資料型態為 category，印出所有欄位的資料型態，並存成新的 CSV 檔 (設定index=False)。(5%)

```python
df.drop(['CustomerId'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Geography'] = df['Geography'].astype('category')
df['Gender'] = df['Gender'].astype('category')
df['HasCrCard'] = df['HasCrCard'].astype('category')
df['Churn'] = df['Churn'].astype('category')
df['IsActiveMember'] = df['IsActiveMember'].astype('category')
```

```python
df.info()
```

```python
df.to_csv("ETC_HW6_107403020.csv", index=False)
```

## (5) 計算屬性間的相關係數，並用 seaborn 繪製出熱力圖 (heatmap)， 如下圖所示。 (8%)

```python
import seaborn as sns
```

```python
columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
tmp = df[columns]
corr = tmp.corr()
plt.figure(figsize=(18, 8))
sns.heatmap(corr, annot=True)
plt.xticks(rotation=0)
plt.show()
```

# 6. 運用資料視覺化來幫助分析
## (1) 繪出 Gender 與 Churn 的數量關係，分析不同性別於客戶流失的關係，如下圖所示。 (Hint: seaborn.countplot()) (10%)

```python
plt.figure(figsize=(16, 6))
ax = sns.countplot(data=df, y='Gender', hue='Churn', palette='Set2')
plt.title('Customer Churn Distribution by Gender')
plt.legend(["Churn", "Not Churn"], title='Churn')
plt.xlabel("count")
plt.ylabel("Gender")
plt.show()
```

```python
gender_group = df.groupby(['Gender', 'Churn']).size().unstack()
gender_group = gender_group[[1,0]]
gender_group.plot(kind='barh', figsize=(8, 6), title='Customer Churn Distribution by Gender')
plt.legend(["Churn", "Not Churn"], title='Churn')
plt.xlabel("count")
plt.ylabel("Gender")
plt.show()
```

## (2) 繪出 Geography 與 Churn 的數量關係，分析不同地區於客戶流失的關係。 (Hint: seaborn.countplot()) (5%)

```python
plt.figure(figsize=(16, 6))
ax = sns.countplot(data=df, y='Geography', hue='Churn', palette='Set2')
plt.title('Customer Churn Distribution by Geography')
plt.legend(["Churn", "Not Churn"], title='Churn')
plt.xlabel("count")
plt.ylabel("Geography")
plt.show()
```

```python
geography_group = df.groupby(['Geography', 'Churn']).size().unstack()
geography_group.plot(kind='barh', figsize=(8, 6), title='Customer Churn Distribution by Geography')
plt.legend(["Not Churn", "Churn"], title='Churn')
plt.xlabel("count")
plt.ylabel("Geography")
plt.show()
```

## (3) 繪出 Age 分布與 Churn 的關係，分析不同年齡於客戶流失率的關係，如下圖所示。 (Hint: seaborn.kdeplot()) (10%)

```python
plt.figure(figsize=(18, 6))
ax = sns.set_palette('Set2')
ax1 = sns.kdeplot(data=df[df['Churn']==0], x='Age', hue='Churn', multiple='stack')
ax2 = sns.kdeplot(data=df[df['Churn']==1], x='Age', hue='Churn', multiple='stack')
plt.title('Customer Age - churn vs. no churn')
plt.legend(["Not Churn", "Churn"])
plt.xlabel('Customer Age')
plt.ylabel('Frequecy')
plt.show()
```

## (4) 繪出 CreditScore 與 Churn 的關係，分析客戶信用分數於客戶流失率的關係。(Hint: seaborn.kdeplot()) (7%)

```python
plt.figure(figsize=(18, 6))
ax = sns.set_palette('Set2')
ax1 = sns.kdeplot(data=df[df['Churn']==0], x='CreditScore', shade=True, hue='Churn')
ax2 = sns.kdeplot(data=df[df['Churn']==1], x='CreditScore', shade=True, hue='Churn')
plt.title('Customer CreditScore - churn vs. no churn')
plt.legend(["Not Churn", "Churn"])
plt.xlabel('Customer CreditScore')
plt.ylabel('Frequecy')
plt.show()
```
