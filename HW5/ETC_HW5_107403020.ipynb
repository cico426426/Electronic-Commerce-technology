# 刪除重複欄位

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

# 填補空值

```python
df['Tenure'] = df.groupby('Gender')['Tenure'].apply(lambda x : x.fillna(x.mean()))
df['WarehouseToHome'] = df['WarehouseToHome'].fillna(df['WarehouseToHome'].mean())
df['HourSpendOnApp'] = df['HourSpendOnApp'].fillna(df['HourSpendOnApp'].mean())
df['OrderAmountHikeFromlastYear'] = df['OrderAmountHikeFromlastYear'].fillna(df['OrderAmountHikeFromlastYear'].mean())
df['CouponUsed'] = df.groupby('Gender')['CouponUsed'].apply(lambda x : x.fillna(x.mean()))
df['OrderCount'] = df.groupby('Gender')['OrderCount'].apply(lambda x : x.fillna(x.mean()))
df['DaySinceLastOrder'] = df.groupby('Gender')['DaySinceLastOrder'].apply(lambda x : x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

# Feature encodeing

```python
df = pd.get_dummies(data=df, columns=['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus'])
df.head()
```

```python
df = df.drop(['Gender_Male'], axis=1)
df.head()
```

# 1. 針對 churn 欄位 使用 Stratified sampling 從原本的資料集中取 60% 的資料

```python
X = df.drop(['Churn'], axis=1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y,random_state=0)
```

# 3. 資料前處理，並以10 folds cross validation 建立 Logistic Regression 及 SVM 模型

```python
from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
cv_Lg_result = cross_validate(LogisticRegression(random_state=0, max_iter=10000), X_train, y_train, cv=10)
cv_Lg_result
```

```python
cv_svm_result = cross_validate( make_pipeline(StandardScaler(), SVC(gamma='auto', random_state=0)), X_train, y_train, cv=10)
cv_svm_result
```

# 4. 針對測試資料 印出兩個模型 的 平均 Accuracy

```python
print("LogisticRegression :",  cv_Lg_result['test_score'].mean())
```

```python
print("SVM : ", cv_svm_result['test_score'].mean())
```

# 5. 重複 1 4 題 3 0 次，並印出 兩種模型 最終 的平均 A ccuracy

```python
def create_data():
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y)
    return X_train, X_test, y_train, y_test
```

```python
def do_LogisticRegression() :
    X_train, X_test, y_train, y_test = create_data()
    cv_Lg_result = cross_validate(LogisticRegression(random_state=0, max_iter=10000), X_train, y_train, cv=10)
    return cv_Lg_result['test_score'].mean()
```

```python
def do_SVM() :
    X_train, X_test, y_train, y_test = create_data()
    cv_svm_result = cross_validate( make_pipeline(StandardScaler(), SVC(gamma='auto', random_state=0)), X_train, y_train, cv=10)
    return cv_svm_result['test_score'].mean()
```

```python
lg_result = []
svm_result = []

for i in range(0, 30):
    lg_result.append(do_LogisticRegression())
    svm_result.append(do_SVM())
```

```python
from statistics import mean

lg_result_mean = mean(lg_result)
svm_result_mean = mean(svm_result)

print("LogisticRegression :",  lg_result_mean)
print("SVM : ", svm_result_mean)
```

# 6. 根據 模型於 30 次 10 folds cross validation 的 Accuracy 以 paired t-test 比較 兩種模型 ，並說明結論

```python
from scipy import stats
import numpy as np
```

```python
lg_np = np.array(lg_result)
svm_np = np.array(svm_result)
```

```python
stats.ttest_rel(lg_np, svm_np)
```

pvalue的值<0.05，代表SVM與LogisticRegression有明顯的差異，應該拒絕LogisticRegression(H0假設)
